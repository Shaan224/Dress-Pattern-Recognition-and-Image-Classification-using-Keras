## Importing Required Libraries

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import urllib.request as request
```

## Loading the CSV file with Images links

```python
dress_data = pd.read_csv("dress_patterns.csv")
```

```python
class_labels = set(dress_data['category'])
```

### Making Image Directory

```python
img_dir=os.mkdir('Image_Datasets')
```

### Making folder for each category

```python

for label in class_labels:
    os.mkdir('Image_Datasets/'+ label)
```

```python
os.chdir(os.getcwd()+'/Image_Datasets')
```

### Downloading the Images from their given Links

```python
%time
for i in range(dress_data.shape[0]):
    try:
        request.urlretrieve(dress_data.iloc[i]["image_url"],os.getcwd()+ '/' + str(dress_data.iloc[i]["category"]) + '/'+ str(dress_data.iloc[i]["_unit_id"])+'.jpg')
    except:
        print("Error in row: ",i)
%time
```
